#Web Scrapping for getting a title of a site

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

```python
URL = "https://www.stirringminds.com/"
```

```python
page = requests.get(URL)
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

```python
requests = soup.findAll(attrs = {'class': 'large-heading-2'})
```

```python
requests
```
